```python
# ============================================================
# ðŸ©º Skin Lesion Hybrid AI Training â€” Final Version
# ============================================================

!pip install -q torch torchvision timm transformers sklearn tqdm matplotlib opencv-python accelerate

import os, random, cv2, timm, torch, pickle
import numpy as np
import torch.nn as nn
import torch.nn.functional as F
from tqdm import tqdm
from PIL import Image
from torch.utils.data import Dataset, DataLoader, random_split
from torchvision import transforms, datasets
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt

# ============================================================
# 1ï¸âƒ£ CONFIGURATION
# ============================================================
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
IMG_SIZE = 224
BATCH_SIZE = 16
CAPTION_EPOCHS = 8
BERT_EPOCHS = 8
LR_CAPTION = 5e-4
LR_BERT = 1e-5
PATIENCE = 3

print(f"âœ… Using device: {DEVICE}")

# ============================================================
# 2ï¸âƒ£ LOAD DATASET (ISIC)
# ============================================================
import kagglehub
print("ðŸ“¥ Downloading ISIC dataset...")
root = kagglehub.dataset_download("nodoubttome/skin-cancer9-classesisic")
DATA_ROOT = os.path.join(root, "Skin cancer ISIC The International Skin Imaging Collaboration", "Train")

train_tfm = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.ColorJitter(0.1, 0.1),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])
])
val_tfm = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])
])

dataset = datasets.ImageFolder(DATA_ROOT, transform=train_tfm)
n_val = int(0.2 * len(dataset))
n_train = len(dataset) - n_val
train_ds, val_ds = random_split(dataset, [n_train, n_val])
val_ds.dataset.transform = val_tfm

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=BATCH_SIZE)

class_names = dataset.classes
num_classes = len(class_names)
print(f"âœ… Dataset: Train={len(train_ds)} Val={len(val_ds)} Classes={num_classes}")

# ============================================================
# 3ï¸âƒ£ ABCDE FEATURE EXTRACTOR
# ============================================================
def extract_abcde(img_tensor):
    img_np = np.transpose(img_tensor.cpu().numpy(), (1,2,0))
    img = (img_np * 255).astype(np.uint8)
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    blur = cv2.GaussianBlur(gray, (5,5), 0)
    mask = cv2.adaptiveThreshold(blur,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                 cv2.THRESH_BINARY_INV,11,2)
    area = np.sum(mask>0) or 1
    diameter_px = np.sqrt(4*area/np.pi)
    diameter_mm = (diameter_px / IMG_SIZE) * 30
    flipped = cv2.flip(mask, 1)
    diff = cv2.absdiff(mask, flipped)
    asym = np.clip(np.sum(diff>0)/(area+1e-6), 0, 1)
    contours,_ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    perim = max(cv2.arcLength(c, True) for c in contours) if contours else 1e-6
    circularity = 4*np.pi*area/(perim**2 + 1e-6)
    border_irreg = np.clip(1-circularity, 0, 1)
    lesion_pix = img[mask>0]
    color_var = np.std(lesion_pix)/255.0 if lesion_pix.size>0 else 0.0
    return {
        "Asymmetry": round(float(asym),3),
        "Border": round(float(border_irreg),3),
        "Color": round(float(color_var),3),
        "Diameter": round(float(diameter_mm),2),
        "Evolution": round(float(color_var),3)
    }

# ============================================================
# 4ï¸âƒ£ CAPTIONING MODEL (EfficientNet + LSTM)
# ============================================================
class CNNEncoder(nn.Module):
    def __init__(self, embed_size=256):
        super().__init__()
        self.cnn = timm.create_model("efficientnet_b0", pretrained=True, num_classes=0)
        self.linear = nn.Linear(self.cnn.num_features, embed_size)
        self.bn = nn.BatchNorm1d(embed_size)
    def forward(self, x):
        x = self.cnn(x)
        x = self.linear(x)
        x = self.bn(x)
        return x

class LSTMDecoder(nn.Module):
    def __init__(self, embed_size, hidden_size, vocab_size):
        super().__init__()
        self.embed = nn.Embedding(vocab_size, embed_size)
        self.lstm = nn.LSTM(embed_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, vocab_size)
    def forward(self, features, captions):
        embeds = self.embed(captions)
        inputs = torch.cat((features.unsqueeze(1), embeds), 1)
        h, _ = self.lstm(inputs)
        outputs = self.fc(h[:,1:,:])
        return outputs

# Vocabulary (basic tokenization)
class Vocabulary:
    def __init__(self):
        self.word2idx = {"<PAD>":0,"<START>":1,"<END>":2,"<UNK>":3}
        self.idx2word = {0:"<PAD>",1:"<START>",2:"<END>",3:"<UNK>"}
        self.idx = 4
    def add_word(self, word):
        if word not in self.word2idx:
            self.word2idx[word] = self.idx
            self.idx2word[self.idx] = word
            self.idx += 1

vocab = Vocabulary()
medical_phrases = [
    "irregular pigmentation", "uniform color", "raised bump",
    "rough scaly texture", "smooth flat lesion", "dark uneven edges",
    "waxy appearance", "brown patch", "asymmetric shape"
]
for phrase in medical_phrases:
    for word in phrase.split():
        vocab.add_word(word)

# Caption model
caption_model = nn.ModuleDict({
    "encoder": CNNEncoder().to(DEVICE),
    "decoder": LSTMDecoder(embed_size=256, hidden_size=512, vocab_size=len(vocab.word2idx)).to(DEVICE)
})

optimizer_cap = torch.optim.AdamW(caption_model.parameters(), lr=LR_CAPTION, weight_decay=1e-4)
criterion_cap = nn.CrossEntropyLoss(ignore_index=0)

# ============================================================
# 5ï¸âƒ£ TRAIN CAPTIONING MODEL
# ============================================================
def generate_caption_template(label):
    return f"lesion showing {random.choice(medical_phrases)} typical of {class_names[label]}"

def prepare_caption_batch(labels):
    caps = []
    for lbl in labels:
        text = "<START> " + generate_caption_template(lbl) + " <END>"
        words = text.lower().split()
        ids = [vocab.word2idx.get(w, 3) for w in words]
        caps.append(ids)
    maxlen = max(len(x) for x in caps)
    pad = np.zeros((len(caps), maxlen))
    for i,seq in enumerate(caps):
        pad[i,:len(seq)] = seq
    return torch.tensor(pad).long()

print("\nðŸŽ¯ Training Captioning Model...")
best_loss = 999; patience = 0
for epoch in range(CAPTION_EPOCHS):
    caption_model.train()
    total = 0
    for imgs, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{CAPTION_EPOCHS}"):
        imgs, labels = imgs.to(DEVICE), labels.to(DEVICE)
        captions = prepare_caption_batch(labels.cpu().numpy()).to(DEVICE)
        features = caption_model["encoder"](imgs)
        outputs = caption_model["decoder"](features, captions[:,:-1])
        loss = criterion_cap(outputs.reshape(-1,len(vocab.word2idx)), captions[:,1:].reshape(-1))
        optimizer_cap.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(caption_model.parameters(), 1.0)
        optimizer_cap.step()
        total += loss.item()
    avg_loss = total/len(train_loader)
    print(f"Epoch {epoch+1} Caption Loss: {avg_loss:.4f}")
    if avg_loss < best_loss:
        best_loss = avg_loss; patience = 0
        torch.save(caption_model.state_dict(), "caption_model.pth")
    else:
        patience += 1
        if patience >= PATIENCE:
            print("â›” Early Stopping Caption Model")
            break

# ============================================================
# 6ï¸âƒ£ TRAIN CLASSIFIER (ClinicalBERT)
# ============================================================
tokenizer = AutoTokenizer.from_pretrained("emilyalsentzer/Bio_ClinicalBERT")
bert_model = AutoModelForSequenceClassification.from_pretrained(
    "emilyalsentzer/Bio_ClinicalBERT", num_labels=num_classes
).to(DEVICE)
optimizer_bert = torch.optim.AdamW(bert_model.parameters(), lr=LR_BERT, weight_decay=1e-4)
criterion_bert = nn.CrossEntropyLoss()

print("\nðŸ§  Training ClinicalBERT on Hybrid Prompts...")
best_acc = 0; patience = 0
train_losses, val_losses, val_accs = [], [], []

for epoch in range(BERT_EPOCHS):
    bert_model.train()
    total_train_loss = 0
    for imgs, labels in tqdm(train_loader, desc=f"BERT Epoch {epoch+1}/{BERT_EPOCHS}"):
        imgs, labels = imgs.to(DEVICE), labels.to(DEVICE)
        features = [extract_abcde(imgs[i]) for i in range(len(imgs))]
        captions = [generate_caption_template(l.item()) for l in labels]
        texts = [
            f"Metrics: {f}. Caption: {c}" for f,c in zip(features, captions)
        ]
        enc = tokenizer(texts, padding=True, truncation=True, return_tensors="pt").to(DEVICE)
        out = bert_model(**enc, labels=labels)
        loss = out.loss
        optimizer_bert.zero_grad(); loss.backward(); optimizer_bert.step()
        total_train_loss += loss.item()
    avg_train = total_train_loss / len(train_loader)
    train_losses.append(avg_train)

    # Validation
    bert_model.eval()
    preds, trues = [], []
    total_val_loss = 0
    with torch.no_grad():
        for imgs, labels in val_loader:
            imgs, labels = imgs.to(DEVICE), labels.to(DEVICE)
            features = [extract_abcde(imgs[i]) for i in range(len(imgs))]
            captions = [generate_caption_template(l.item()) for l in labels]
            texts = [f"Metrics: {f}. Caption: {c}" for f,c in zip(features, captions)]
            enc = tokenizer(texts, padding=True, truncation=True, return_tensors="pt").to(DEVICE)
            out = bert_model(**enc, labels=labels)
            total_val_loss += out.loss.item()
            preds.extend(out.logits.argmax(1).cpu().numpy())
            trues.extend(labels.cpu().numpy())
    avg_val = total_val_loss / len(val_loader)
    acc = accuracy_score(trues, preds)
    val_losses.append(avg_val)
    val_accs.append(acc)
    print(f"Epoch {epoch+1} Loss={avg_train:.4f} | ValLoss={avg_val:.4f} | ValAcc={acc*100:.2f}%")

    if acc > best_acc:
        best_acc = acc; patience = 0
        bert_model.save_pretrained("final_best_hybrid_bert")
        tokenizer.save_pretrained("final_best_hybrid_bert")
    else:
        patience += 1
        if patience >= PATIENCE:
            print("â›” Early Stopping ClinicalBERT")
            break

# ============================================================
# 7ï¸âƒ£ SAVE FINAL COMPONENTS
# ============================================================
torch.save({
    "encoder_state": caption_model["encoder"].state_dict(),
    "decoder_state": caption_model["decoder"].state_dict(),
    "vocab": vocab.word2idx,
    "class_names": class_names
}, "captioning_and_meta.pth")

print(f"\nâœ… Training complete! Best Val Acc: {best_acc*100:.2f}%")
print("ðŸ“¦ Models saved: captioning_and_meta.pth & final_best_hybrid_bert/")

# ============================================================
# 8ï¸âƒ£ PLOT RESULTS
# ============================================================
plt.figure(figsize=(7,4))
plt.plot(train_losses, label="Train Loss")
plt.plot(val_losses, label="Validation Loss")
plt.title("ClinicalBERT Training Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend(); plt.grid(True)
plt.show()

plt.figure(figsize=(7,4))
plt.plot(val_accs, label="Validation Accuracy", marker='o')
plt.title("Validation Accuracy per Epoch")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(); plt.grid(True)
plt.show()
```
